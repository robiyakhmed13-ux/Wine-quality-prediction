Data Preprocessing

```python
#Labeling and Separate the data
X = wine_dataset.drop(columns='quality', axis=1)
```

Laber Binarization

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value >=7 else 0)
```

```python
print(Y)
```

Training & Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, 
    Y,
    test_size = 0.2,
    stratify=Y,
    random_state=2
)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, X.shape)
```

Building the model

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

Evaluating the model

```python
training_data_predcition = model.predict(X_train)
training_accuracy_score = accuracy_score(training_data_predcition,Y_train )
print('Accuracy score:', training_accuracy_score)
```

```python
test_data_predcition = model.predict(X_test)
test_accuracy_score = accuracy_score(test_data_predcition, Y_test)
print('Accuracy score:', test_accuracy_score)
```

Making predictive system

```python
input_data = (7.5,0.5,0.36,6.1,0.071,17.0,102.0,0.9978,3.35,0.8,10.5)

input_data_numpy = np.asarray(input_data)
input_data_reshaped = input_data_numpy.reshape(1, -1)
prediction = model.predict(input_data_reshaped)

```

```python
if prediction[0] == 0:
    print('wine is high quality')
else:
    print('Wine is not high quality')
```
